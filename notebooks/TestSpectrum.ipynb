### Making a flux spectrum

Often, we'd like to plot the results of the analysis in flux units, rather than the detector counts units. There is now code in Clarsach to do that.

**Warning**: In practice, getting the flux spectrum out of an observation in detector space is a deconvolution operation, which is hard. The `calculate_flux_spectrum` function provides an approximation that should *not* be used for any data analysis, but for plotting purposes only.

```python
from clarsach import calculate_flux_spectrum
```

```python
flux_spec = calculate_flux_spectrum(spec)
```

```python
ax = plt.subplot(111)
ax.loglog(spec.bin_mid, flux_spec, color='r', alpha=0.8)
plt.title(testfile)
```

```python
ax = plt.subplot(111)
spec.plot(ax, xunit='keV', yunit="flux", color='r', alpha=0.8)
plt.title(testfile)
ax.set_xscale("log")
ax.set_yscale('log')
```
